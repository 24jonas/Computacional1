```python
#Importamos bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Leemos archivos
df1 = pd.read_csv('AntarticaJunio22.txt', skiprows=10, sep='\\s+')
df2 = pd.read_csv('AntarticaDiciembre22.txt', skiprows=10, sep='\\s+')
df1.columns = ['PRES', 'HGHT', 'TEMP', 'DWPT', 'RELH', 'MIXR', 'DRCT', 'SKNT', 'THTA', 'THTE', 'THTV']
df2.columns = ['PRES', 'HGHT', 'TEMP', 'DWPT', 'RELH', 'MIXR', 'DRCT', 'SKNT', 'THTA', 'THTE', 'THTV']
```

```python
#Graficamos presión contra altura de Junio
df01 = df1[['HGHT','PRES']]
plt.figure(); df01.plot(x='HGHT'); plt.legend(loc='best')
plt.title('Variación de la presión contra altura Junio')
plt.ylabel('Presión hPa')
plt.xlabel('Altura m')
plt.grid(True)
plt.show()
```

```python
#Graficamos presión contra altura de Diciembre
df02 = df2[['HGHT','PRES']]
plt.figure(); df02.plot(x='HGHT'); plt.legend(loc='best')
plt.xlim((0,34000))
plt.title('Variación de la presión contra altura Diciembre')
plt.ylabel('Presión hPa')
plt.xlabel('Altura m')
plt.grid(True)
plt.show()
```

```python
#Graficamos temperatura contra altura de Junio
df03 = df1[['HGHT','TEMP']]
plt.figure(); df03.plot(x='HGHT'); plt.legend(loc='best')
plt.title('Variación de la temperatura contra altura')
plt.ylabel('Temperatura C')
plt.xlabel('Altura m')
plt.grid(True)
plt.show()
```

```python
#Graficamos temperatura contra altura de Diciembre
df04 = df2[['HGHT','TEMP']]
plt.figure(); df04.plot(x='HGHT'); plt.legend(loc='best')
plt.xlim((0,34000))
plt.title('Variación de la temperatura contra altura Diciembre')
plt.ylabel('Temperatura C')
plt.xlabel('Altura m')
plt.grid(True)
plt.show()
```

```python
#Graficamos temperatura y temperatura de rocio contra altura de Junio
df05 = df1[['HGHT','TEMP','DWPT']]
plt.figure(); df05.plot(x='HGHT'); plt.legend(loc='best')
plt.title('Variación de la temperatura y temp. de rocio contra altura Junio')
plt.ylabel('Temperatura C')
plt.xlabel('Altura m')
plt.grid(True)
plt.show()
```

```python
#Graficamos temperatura y temperatura de rocio contra altura de Diciembre
df06 = df2[['HGHT','TEMP','DWPT']]
plt.figure(); df06.plot(x='HGHT'); plt.legend(loc='best')
plt.xlim((0,34000))
plt.title('Variación de la temperatura y temp. de rocio contra altura Diciembre')
plt.ylabel('Temperatura C')
plt.xlabel('Altura m')
plt.grid(True)
plt.show()
```

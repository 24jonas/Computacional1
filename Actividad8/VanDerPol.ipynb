```python
# Importa bibliotecas
from scipy.integrate import odeint 
from scipy import array, arange,pi, sin
import matplotlib.pyplot as plt

# Define las ecuaciones diferenciales del oscilador de Van del Pol 
def VanDerPol(X,t):
    x = X[0]
    x_point = X[1]
    dx = x_point
    dx_point = epsilon*(1 - x*x)*dx - x
    return array([dx, dx_point])
```

```python
# Valores de parámetros
t0 = 10
tmax = 50
pastemps = 0.01
time = arange(t0, tmax, pastemps)

epsilon = 5.0

# Condiciones iniciales
x0 = 2.0
v0 = 0.0

# Se crean muestras de tiempo
x, x_point = odeint(VanDerPol,(x0,v0),time).T

# Gráfica de posición contra tiempo
p1 = plt.figure()
plt.grid()
plt.xlabel('$t$', fontsize = 16)
plt.ylabel('$x$', fontsize = 16)
plt.plot(time, x)
plt.savefig('xtsinf.png')
plt.show()
```

```python
import numpy as np
# Define las ecuaciones diferenciales del oscilador de Van del Pol 
def VanDerPol(X,t):
    x = X[0]
    x_point = X[1]
    dx = x_point
    dx_point = epsilon*(1 - x*x)*dx - x + A * (np.sin(w * t))
    return array([dx, dx_point])
```

```python
# Valores de parámetros-------------------------------------------------------------
t0 = 0
tmax = 600
pastemps = 0.01
time = arange(t0, tmax, pastemps)

epsilon = 8.53
w = (2 * 3.14159265)/10
A = 1.2

# Condiciones iniciales
x0 = 0.0
v0 = 2.0

# Se crean muestras de tiempo
x, x_point = odeint(VanDerPol,(x0,v0),time).T

with open('xtconf.dat', 'w') as f:
    # Print & save the solution.
    for t1, x1, x_point1 in zip(time, x, x_point):
        print (t1, x1, x_point1, file=f)

# Gráfica de fase-------------------------------------------------------------------

# Plot the solution that was generated\n",
from numpy import loadtxt
from pylab import figure, plot, xlabel, ylabel, grid, hold, legend, title, savefig
from matplotlib.font_manager import FontProperties
%matplotlib inline

t1, x, y = loadtxt('xtconf.dat', unpack=True, skiprows=30000)
figure(1, figsize=(6, 4.5))
xlabel('t')
grid(True)
lw = 0.5
plot(t1, x, 'r', linewidth=lw)
title('Oscilador de Van der Pol con forzamiento sinusoidal')
savefig('xtconf.png', dpi=100)
```
